```python
import spacy
nlp = spacy.load('/opt/anaconda3/lib/python3.7/site-packages/en_core_web_lg/en_core_web_lg-2.2.5/')

import pandas as pd
import numpy as np
import re
```

```python
climate = open("/Users/shiqinchoo/Desktop/combined/climate.txt","r")
climate = climate.read()

bio = open("/Users/shiqinchoo/Desktop/combined/bio.txt","r")
bio = bio.read()

connect = open("/Users/shiqinchoo/Desktop/combined/connect.txt","r")
connect = connect.read()

trade = open("/Users/shiqinchoo/Desktop/combined/trade.txt","r")
trade = trade.read()

jobs = open("/Users/shiqinchoo/Desktop/combined/jobs.txt","r")
jobs = jobs.read()

social = open("/Users/shiqinchoo/Desktop/combined/social.txt","r")
social = social.read()
```

```python
def ent_count(text,name):
    
    ents = []
    doc = nlp(text, disable=['parser']) #remove 'disable' if not more than 1000000 characters
    ent_list = [ent.text for ent in doc.ents]
    ents.append(ent_list)

    print(len(ents))
    
    ent_lab = []

    lab = [ent.label_ for ent in doc.ents]
    ent_lab.append(lab)
    
    print(len(lab))
    
    entities = []
    for sublist in ents:
        for item in sublist:
            entities.append(item)
            
    print(len(entities))

    labels = []
    for sublist in ent_lab:
        for item in sublist:
            labels.append(item)
        
    print(len(labels))
    
    ent_count = pd.DataFrame()
    ent_count['entities'] = entities
    ent_count['labels'] = labels
    
    ent_size = ent_count.groupby(['entities','labels']).size()
    ent_size = ent_size.to_frame(name='size').reset_index()
    ent_size = ent_size.sort_values(by='size',ascending=False).reset_index()
    
    ent_size.to_csv(name +'.csv')
    return ent_size
    
    
```
